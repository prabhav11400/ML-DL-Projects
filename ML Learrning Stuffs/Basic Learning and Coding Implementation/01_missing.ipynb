Methods to handle missing values: Imputation and Dropping

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dropping = drop all the rows with any missing values. Useful when 1M, 1B itne jyadya , too much rows/datasets hon
```

```python
# Imputation : filling with central tendencies:
# 1. mean
# 2. median
# 3. mode
```

Instead of dropping full rows, imputation is better choice

```python
# analyse the distribution of data in salary column
# fig,ax = plt.subplots(figsize=(8,8))
# sns.distplot(dataset.salary) : don't use this alternatives use kro
```

```python
#https://gist.github.com/mwaskom/de44147ed2974457ad6372750bbe5751

sns.histplot(dataset["salary"], kde=True)
```

```python

sns.histplot(dataset["salary"], kde=True, stat="density")
```

we clearly see, outliers. Skewed distribution : so going with mean replacement will not be a good imputation techniques. OUTLIERS AND SKEWED : no means use.
Similar to normal distribution ke case me use : mean

```python
# Replacing the miising values with median/middle value : median is beeter choice when there are outliers
```

```python
dataset['salary'].fillna(dataset['salary'].median(),inplace=True)
```

```python
dataset.isnull().sum() # now number of missing values in each column 0
```

```python
sns.histplot(dataset["salary"], kde=True, stat="density")
```

```python
# when normal distribution mila hota check krne pr
# filling with mean
dataset['salary'].fillna(dataset['salary'].mean(),inplace=True)
```

```python
sns.histplot(dataset["salary"], kde=True, stat="density")
```

```python
# filling with mode
dataset['salary'].fillna(dataset['salary'].mode(),inplace=True)
```

```python
sns.histplot(dataset["salary"], kde=True, stat="density")
```

```python
# Dropping method : 215 me se 67 rows will be deleted, that is not good as dataset chota hai
# Bade dataset me drop kiya karo
# loading the dataset to a pandas dataframe
dataset=pd.read_csv('/content/Placement_Dataset.csv')
```

```python
dataset.isnull().sum()
```

```python
sns.histplot(dataset["salary"], kde=True, stat="density")
```

```python
dropped=dataset.dropna(how='any')
```

```python
dropped.isnull().sum()
```

```python
dropped.shape # 215-67 = 148
```
