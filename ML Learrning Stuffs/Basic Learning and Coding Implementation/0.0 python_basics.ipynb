# Functions in pyhton

```python
n=int(input("Enter the number jiska factorial chahiye: "))
def rec_fact(n):
  if ( n==0 or n==1 ) :
    return 1;
  return n*rec_fact(n-1)
print(rec_fact(n))
```

```python
def fact(n):
  if(n==0 or n==1):
    return 1
  else :
    factorial=1
    for i in range(1,n+1):
      factorial=factorial*i
    return factorial
num=int(input())
print(fact(num)) # iteratively max 1558 ka factorial, recursively max 959 ka colab me, vs code recursively in my pc 998 max, iterative vs code max 1558 same as colab
```
