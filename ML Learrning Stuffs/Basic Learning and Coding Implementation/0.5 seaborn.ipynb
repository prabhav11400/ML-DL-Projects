## Creating BOSTON HOUSE DATASET FROM SCRATCH

```python
# house prize dataset : sklearn se ethical issue so use alternate way
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]

raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None, names=column_names)

```

```python
# from raw_df dataframe, I want to drop firstly column whose name is 'B', as all NONE hai
raw_df=raw_df.drop(columns='B')
# alternate : raw_df.drop(columns='B', inplace=True) bina equate kiye
```

```python
raw_df
```

```python
#
# now i want to drop every odd indexed rows as odd indexed rows me first 2 column ke bad all none hain
raw_df=raw_df.iloc[::2] # Select every even-indexed row
```

```python
raw_df
```

```python
# So finally I have successfully created the raw_df or boston house prize dataset using chatgpt and sklearn documentation. Hope link pr se bhi data ko nhi hatayenge.
```

```python
raw_df.head()
```

```python
# still LSTAT	MEDV are none, drop them to
raw_df=raw_df.drop(columns=['LSTAT','MEDV'])
```

```python
raw_df
```

```python
# resetting the indexes again
raw_df = raw_df.reset_index(drop=True)
```

```python
raw_df
```

```python
# this data/dataframe does not contain the price of the houses array.
# adding price column by the help of machine learning, by analysing already available collumns
# but machine learning se nhi kr skte, as we don't know aven a single house prize value
# machine learning toh kuch available data ko analyse krke kuch bachi ko predict krta hai.
# But hamare pas toh kuch bhi nhi hai analyse krne ke liye
```

```python
# one way is to add diectly by
# Assuming you have already imported your 'raw_df' DataFrame

# Create a synthetic 'PRICE' column based on a linear combination of other columns
raw_df['PRICE'] = (
    25 +                   # Intercept
    10 * raw_df['RM'] +     # Number of rooms
    3 * raw_df['ZN'] -      # Zoning proportion
    5 * raw_df['CRIM'] +    # Crime rate, km honi chahiye, mahnge hose ke liye
    2 * raw_df['TAX'] -     # Property tax
    3 * raw_df['PTRATIO']     # pupil-teacher ratio km = better education, assumption
)

# Display the updated DataFrame with the synthetic 'PRICE' column
print(raw_df)
```

```python
# finally approx similar boston house dataset I have created successfully
```

```python
raw_df.head()
```

```python
# B and LSTAT should also be there in my dataframe. Let's add them too.
# Create synthetic 'B' and 'LSTAT' columns based on other columns
raw_df['B'] = raw_df['CRIM'] * 100 + raw_df['ZN'] * 10  # Synthetic formula for 'B'
raw_df['LSTAT'] = raw_df['AGE'] / 10 + raw_df['INDUS'] * 5  # Synthetic formula for 'LSTAT'
```

```python
raw_df.head()
```

```python
# Divide the 'PRICE' column by 20
raw_df['PRICE'] = raw_df['PRICE'] / 20 # so that original ke similar dikhe
```

```python
raw_df
```

```python
# To shift the 'PRICE' column to the last position in your DataFrame, you can use the following
# Get the column names in the current order
columns = raw_df.columns.tolist()

# Move 'PRICE' to the last position
columns.remove('PRICE')
columns.append('PRICE')

# Reorder the DataFrame columns
raw_df = raw_df[columns]
```
