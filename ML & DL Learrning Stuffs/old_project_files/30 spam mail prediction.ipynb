```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# replace the null values with a null string
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```


The code you provided seems to be attempting to replace NaN or missing values in a DataFrame with empty strings. In recent versions of pandas, the .where() method is used slightly differently, especially with the replacement argument.

In this code, you are using the pd.notnull(raw_mail_data) as a mask to identify where values are not null (i.e., not NaN), and where this condition is True, it keeps the original value. In cases where it's False, it replaces the value with an empty string ''.

```python
# labeling spam mails as 0, ham mail as 1
# mail_data.loc[mail_data['Category']=='spam', 'Category'] = 0
# mail_data.loc[mail_data['Category']=='ham', 'Category']=1
# alternative and simple form of the above two lines
mail_data['Category'] = mail_data['Category'].replace({'spam': 0, 'ham': 1})
```

```python
X = mail_data['Message']
Y = mail_data['Category']
```

```python
# Splitting the data into training and testing data
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,stratify=Y, test_size=0.33, random_state=4)
```

Feature extraction

```python
# transform the text data into feature vectors that can be used as input to the logistic regression
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english',lowercase=True)
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test) # fit_Transform only train pr, test pr fit nhi krna
```

```python
print(x_train_features)
```

```python
# machine understand numbers not texts. So we will use x_train_feature from now
```

Training the model

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with the training data
model.fit(x_train_features, y_train)
```

```python
# Evaluating the model
```

```python
# Prediction on the training data
```

```python
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
accuracy_on_training_data
```

```python
prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test, prediction_on_test_data)
```

```python
accuracy_on_test_data
```

```python
# clearly no overfitting in the model
# since accuracy on training and testing are almost same, not much difference
```
