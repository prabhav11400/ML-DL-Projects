In machine learning, data standardization is a process of transforming all the data to have the same scale or range. This is important because many machine learning algorithms are sensitive to the scale of the input features. When people talk about the order in which to perform data standardization and data splitting for training and testing, there are two common approaches:


Standardize First, Then Split:
In this approach, you standardize the entire dataset, which means you adjust the values so that they all fall within a common range. This ensures that the data is consistent and easier for the machine learning model to work with.
After standardization, you split the dataset into a training set and a testing set. The training set is used to train the machine learning model, and the testing set is used to evaluate its performance.
The advantage of this approach is that it retains the original range of the dataset. This can be important if you need to interpret or visualize the data in a way that makes sense in the original units of measurement.


Split First, Then Standardize:
In this approach, you split the dataset into a training set and a testing set first, and then you standardize each set separately. This means that you adjust the scale of the training and testing data independently.
The advantage of this approach is that it simulates a more realistic scenario where your model encounters unseen data that hasn't been standardized with the training data. It can help prevent information leakage from the testing set to the training set.
However, one potential drawback is that the range of the data may differ between the training and testing sets, which can affect the model's performance.
In simple terms, the first approach (standardize first, then split) makes sure all your data is on the same scale before you train your model. This is useful if you want to work with a consistent dataset, and it helps ensure that your model's behavior is based on the properties of the data rather than its scale.

The second approach (split first, then standardize) aims to mimic a more realistic scenario where your model is applied to unseen data. This approach might lead to slight differences in the range of values between your training and testing data. While it's useful for avoiding information leakage, it's important to be aware of the potential differences in data range and their impact on your model.
