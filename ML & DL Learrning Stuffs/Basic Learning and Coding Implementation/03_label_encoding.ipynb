Label Encoding


*   Converting the labels into numeric form



```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# loading the data from csv to panda
cancer_data=pd.read_csv("/content/data.csv")
# data.csv is nothing but 'breast_cancer_dataset' only
```

```python
cancer_data.head() # features= parameters for cancer cells
```

```python
# finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
# converting M, B to numeric. As analysing numerical is easier than Characters
```

```python
#Loading the label encoder function
label_encode=LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
cancer_data['diagnosis'].value_counts()
```

```python
cancer_data['target'].value_counts()
```

```python
# from the above two codes it is clear that 0 = B, 1 = M
# kon sa label (A,B,C..) ko kon si value(0,1,2,...) milegi
# this is decided by Arranging the labels A,B,C,.. in increasing or alphabetical order. So A =0, B=1, C=3; if Labels A, B C teeno present hon
```
