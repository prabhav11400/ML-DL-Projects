Separating the data : legit and fraudulent transections

```python
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

# Undersampling

Build a sample dataset containing similar distribution of legit and fradulent transections

```python
legit_sample=legit.sample(n=492) # 492 random values from legit df, will be sampled
```

```python
legit_sample.shape
```

Concatenate the two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0) # axis=0 : row wise or top on the other concatenate
```

```python
new_dataset # sare 0 first then sare 1's at last (Class dekho)
```
