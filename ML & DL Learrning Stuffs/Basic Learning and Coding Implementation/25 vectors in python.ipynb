# vector operations in python : Dot, Cross, Projection

dot product

```python
a=np.array([2,3])
b=np.array([4,4])
a_dot_b = np.dot(a,b)
a_dot_b
```

```python

c=np.array([40,50,60])
d=np.array([54,23,44])
print(np.dot(c,d))
```

Cross Product of two vectors

```python
a=np.array([2,3,4])
b=np.array([4,5,6])
a_cross_b=np.cross(a,b)
print(a_cross_b) # i(.-.)-j(.-.)+k(.-.)
```

Projection of one vector onto another vector

```python
# a ka b pr progection = a.cos(0) = a.[(a.b)/|a||b| ] ; a=|a|*(unit vector in direction of a)
```

```python
a=np.array([2,5]) # by default position vectors mano i.e. 0,0,2,5
v=np.array([8,-6])
magnitude_of_v=np.sqrt(np.sum(v**2))
proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print(proj_of_a_on_v)
```

```python
# three dimensional
a=np.array([23,45,62]) # by default position vectors mano i.e. 0,0,2,5
v=np.array([45,82,67])
magnitude_of_v=np.sqrt(np.sum(v**2))
proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v
print(proj_of_a_on_v)
```
