```python
# create two matrices with random values
matrix_1 = np.random.randint(10,size=(3,3))
matrix_2 = np.random.randint(20,size=(3,3))
```

```python
print(matrix_1, matrix_2)
```

```python
sum = np.add(matrix_1,matrix_2) # alternate of +
print(sum)
```

```python
print(np.subtract(matrix_1,matrix_2))
```

```python
diff_matrix = matrix_1 - matrix_2
print(diff_matrix)
```

```python
# scaler multiplication
x=5
grid=np.random.randint(5,size=(5,6))
grid
```

```python
scaler_multiplication = np.multiply(x,grid)
print(scaler_multiplication)
```

Matrix Multiplication

```python
A = np.random.randint(5, size=(3,4))
B = np.random.randint(5, size=(4,2))
# A*B will be of size 3*2
```

```python
print(A)
```

```python
print(B)
```

```python
product = np.dot(A,B) # A*B
print(product)
```
