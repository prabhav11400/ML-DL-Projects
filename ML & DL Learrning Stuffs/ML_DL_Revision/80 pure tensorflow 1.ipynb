In the context of deep learning and machine learning frameworks, tensors are multi-dimensional arrays that can represent data. Initialization of tensors involves assigning initial values to the elements of the tensor. Proper initialization is crucial for the training process as it can affect the convergence and performance of the model.

Here's a brief overview of how tensors are typically initialized in popular deep learning frameworks such as TensorFlow and PyTorch:

### TensorFlow:

In TensorFlow, you can use the `tf.constant`, `tf.Variable`, or specific initialization functions from the `tf.initializers` module.

1. **Constant Initialization:**
   ```python
   import tensorflow as tf

   # Initialize a constant tensor
   tensor_constant = tf.constant([1, 2, 3])

   # Initialize a variable with a constant value
   variable_constant = tf.Variable([1, 2, 3])
   ```

2. **Random Initialization:**
   ```python
   # Initialize a variable with random values using Glorot uniform initialization
   variable_random = tf.Variable(tf.initializers.glorot_uniform()([3, 3]))
   ```

### PyTorch:

In PyTorch, tensors are typically initialized using functions from the `torch` module.

1. **Constant Initialization:**
   ```python
   import torch

   # Initialize a constant tensor
   tensor_constant = torch.tensor([1, 2, 3])

   # Initialize a variable with a constant value
   variable_constant = torch.nn.Parameter(torch.tensor([1, 2, 3]))
   ```

2. **Random Initialization:**
   ```python
   # Initialize a variable with random values using Xavier/Glorot uniform initialization
   variable_random = torch.nn.Parameter(torch.nn.init.xavier_uniform_(torch.empty(3, 3)))
   ```

The code `tf.random.normal((2, 3))` is an example of using TensorFlow to create a tensor with random values drawn from a normal distribution.

Here's a breakdown:

```python
import tensorflow as tf

# Create a tensor with random values from a normal distribution
random_tensor = tf.random.normal((2, 3))
```

In this example:
- `tf.random.normal` is a TensorFlow function that generates a tensor with values sampled from a normal (Gaussian) distribution.
- The argument `(2, 3)` specifies the shape of the tensor. In this case, it's a 2D tensor with 2 rows and 3 columns.

The values in the `random_tensor` will follow a normal distribution with a mean of 0 and a standard deviation of 1 by default, but you can customize the mean and standard deviation by providing additional arguments to `tf.random.normal`. For example:

```python
# Create a tensor with random values from a normal distribution
custom_random_tensor = tf.random.normal((2, 3), mean=0, stddev=1.5)
```

Here, `mean` sets the mean of the normal distribution, and `stddev` sets the standard deviation. Adjusting these parameters allows you to control the characteristics of the random values generated in the tensor.

# Creating basic neural network using sequential api of keras

Sequential API = very convenient, not very flexible. = only allows you to have one input mapped to one output.

**EXPLANATION:**The Sequential API in deep learning frameworks like TensorFlow and Keras is designed to be convenient and easy to use, especially for building simple feedforward neural networks where the data flows sequentially through the layers. However, its simplicity comes with certain limitations in terms of flexibility.

Here's an explanation of the statement:

1. **Convenient but Less Flexible:**
   - **Convenient:** The Sequential API is convenient because it allows you to create a neural network layer by layer in a linear, sequential fashion. You add one layer at a time, and it's easy to understand the flow of data.
   - **Less Flexible:** However, this simplicity comes at the cost of flexibility. With the Sequential API, you can only create models where each layer has one input tensor and produces one output tensor. This limitation makes it less suitable for building complex models with multiple inputs or outputs, shared layers, or architectures with more intricate connections.

2. **One Input to One Output Mapping:**
   - In the Sequential API, you add layers one by one, and each layer takes the output of the previous layer as its input. This sequential flow enforces a mapping of one input to one output.
   - If you have a model architecture where there are multiple inputs or outputs, or if you need to create more complex connections between layers (e.g., shared layers), the Sequential API may not be sufficient, and you'd need to use the more flexible Functional API.

3. **Example of Limitation:**
   - For instance, imagine a scenario where you want to create a model with two separate input branches that later merge before producing the final output. Achieving this with the Sequential API becomes challenging because it doesn't inherently support such complex topologies.

In contrast to the Sequential API, the Functional API in Keras or TensorFlow provides more flexibility. It allows you to create models with multiple inputs, multiple outputs, shared layers, and non-linear connections between layers. The trade-off is that the Functional API is slightly more complex than the Sequential API.

In summary, the Sequential API is a great choice for simple, linear models, while the Functional API provides more flexibility for building complex and non-sequential architectures. The choice depends on the complexity of your model and the specific requirements of your deep learning task.

# Functional API : A bit more flexible(multiple inputs, multiple outputs)

The Functional API in Keras (and TensorFlow) provides a more flexible way to define and build neural network models compared to the Sequential API. It allows you to create models with multiple inputs, multiple outputs, shared layers, and complex network architectures. Here's an explanation of its flexibility:

### 1. **Multiple Inputs:**
   - With the Functional API, you can easily create models that take multiple inputs. This is useful in scenarios where your model needs to process different types of data simultaneously.
   - Example:
     ```python
     from tensorflow.keras.layers import Input, Dense, concatenate

     # Define two input layers
     input1 = Input(shape=(64,))
     input2 = Input(shape=(32,))

     # Merge inputs using a concatenation layer
     merged = concatenate([input1, input2])

     # Add more layers to the merged input
     hidden_layer = Dense(128, activation='relu')(merged)
     output = Dense(1, activation='sigmoid')(hidden_layer)

     # Create the model
     model = keras.Model(inputs=[input1, input2], outputs=output)
     ```

### 2. **Multiple Outputs:**
   - The Functional API supports models with multiple outputs, which can be useful in tasks like multi-task learning.
   - Example:
     ```python
     from tensorflow.keras.layers import Input, Dense

     # Define an input layer
     input_data = Input(shape=(10,))

     # Create two branches of the network
     branch1 = Dense(64, activation='relu')(input_data)
     branch2 = Dense(32, activation='relu')(input_data)

     # Concatenate the branches
     merged = concatenate([branch1, branch2])

     # Additional layers for the merged data
     output1 = Dense(1, activation='sigmoid', name='output1')(merged)
     output2 = Dense(10, activation='softmax', name='output2')(merged)

     # Create the model with multiple outputs
     model = keras.Model(inputs=input_data, outputs=[output1, output2])
     ```

### 3. **Shared Layers:**
   - You can easily share layers between different parts of your model using the Functional API. This is beneficial when you want to reuse the same layer for different inputs or branches.
   - Example:
     ```python
     from tensorflow.keras.layers import Input, Dense

     # Define an input layer
     input_data = Input(shape=(10,))

     # Shared layer
     shared_layer = Dense(64, activation='relu')

     # Two branches sharing the same layer
     branch1 = shared_layer(input_data)
     branch2 = shared_layer(input_data)

     # Additional layers for each branch
     output1 = Dense(1, activation='sigmoid', name='output1')(branch1)
     output2 = Dense(10, activation='softmax', name='output2')(branch2)

     # Create the model
     model = keras.Model(inputs=input_data, outputs=[output1, output2])
     ```

### 4. **Complex Architectures:**
   - The Functional API allows you to create complex model architectures with intricate connections between layers. You have more control over how layers are connected compared to the Sequential API.
   - Example:
     ```python
     from tensorflow.keras.layers import Input, Dense, concatenate

     # Define multiple input layers
     input1 = Input(shape=(64,))
     input2 = Input(shape=(32,))

     # Branch 1
     branch1 = Dense(64, activation='relu')(input1)  # input 1, is the input of branch 1
     branch1 = Dense(32, activation='relu')(branch1) # upar ki branch 1 is now input of branch 1, which makes the modified branch 1

     # Branch 2
     branch2 = Dense(32, activation='relu')(input2)
     branch2 = Dense(16, activation='relu')(branch2)

     # Merge branches
     merged = concatenate([branch1, branch2])

     # Additional layers for the merged data
     output = Dense(1, activation='sigmoid')(merged)

     # Create the model
     model = keras.Model(inputs=[input1, input2], outputs=output)
     ```

In summary, the Functional API provides a more versatile way to construct neural network models, making it suitable for a wider range of tasks and model architectures compared to the Sequential API. It offers greater flexibility and control over the connections between layers, making it a preferred choice for building more complex models.

**Recap CNN**

Certainly! Let's go through the key concepts of Convolutional Neural Networks (CNNs) in a brief manner:

### 1. Convolutional Layer:
   - **Padding:**
     - **"Valid" Padding:** No padding is added to the input image. As a result, the output size decreases.
     - **"Same" Padding:** Zero-padding is added to the input image to maintain the spatial dimensions in the output.

   - **Kernel Size:**
     - The size of the convolutional filter or kernel. It is usually a small matrix applied to the input image to extract features.

   - **Strides:**
     - The step size with which the kernel moves across the input image. Larger strides result in smaller output sizes.

### 2. Pooling Layer:
   - **Pooling:**
     - Pooling layers (e.g., MaxPooling or AveragePooling) downsample the spatial dimensions of the input, reducing computational complexity.
     - Common pooling sizes are 2x2, and the operation (max or average) is applied to each region.

### 3. Flattening Layer:
   - **Flattening:**
     - After convolutional and pooling layers, the data is flattened into a 1D vector before being fed into fully connected layers.

### 4. Fully Connected (FC) Layers:
   - **Fully Connected Layers:**
     - These are traditional neural network layers where each neuron is connected to every neuron in the previous and next layers.
     - Often used after flattening to make predictions or classifications.

### Workflow Summary:
1. **Convolutional Layers:**
   - Detect features using convolutional filters.(loacally mapping)
   - Activation functions like ReLU introduce non-linearity.
   - Choose padding ("valid" or "same"), kernel size, and strides.

2. **Pooling Layers:**
   - Downsample the spatial dimensions.
   - Reduce computational complexity.
   - Commonly use MaxPooling or AveragePooling.

3. **Flattening:**
   - Convert the 3D output to a 1D vector.

4. **Fully Connected (FC) Layers:**
   - Traditional neural network layers.
   - Used for making predictions or classifications.

### Example CNN Architecture:
```plaintext
Input Image -> Convolutional Layer -> Pooling Layer -> Convolutional Layer -> Pooling Layer -> Flattening -> Fully Connected Layers -> Output
```

Remember, the specific hyperparameters (filter size, strides, pooling type, etc.) depend on the problem, and experimentation is often required to find the best configuration. The convolutional and pooling layers are responsible for learning hierarchical features from the input data, while fully connected layers combine these features for final predictions.

2. In a convolutional neural network (CNN), a filter (also referred to as a kernel) is a small matrix that is convolved with the input data (such as an image) to produce a feature map. 3. In a convolutional layer, each filter is convolved with the input data, and each filter produces a single feature map.
````python
def my_model():
  inputs = keras.Input(shape=(32,32,3))
  x = layers.Conv2D(32,3)(inputs) # 32 filters(32 features khojega, 3*3 = kernel size)
  x = layers.BatchNormalization()(x)
  x = keras.activations.relu(x)
  x = layers.MaxPooling2D()(x)
  x = layers.Conv2D(64,5,padding='same')(x)
  x = layers.BatchNormalization()(x)
  x = keras.activations.relu(x)
  x = layers.Conv2D(128,3)(x)
  x = layers.BatchNormalization()(x)
  x = layers.Flatten()(x)
  x = layers.Dense(64, activation='relu')(x) # yhan activation hai, s nhi laga
  outputs = layers.Dense(10)(x)

  model = keras.Model(inputs=inputs, outputs=outputs) # main line to create functional model
  return model

# bina function banaye bhi kr skte the, as we have done before
````

Batch Normalization normalizes the inputs of a neural network layer during training, stabilizing and accelerating training by reducing internal covariate shift, enabling higher learning rates, and acting as a form of regularization.

````python
model = my_model()

model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True), # upar output layer me softmax use nahi kiya so from_logit ko True kiya
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)

model.fit(x_train, y_train, batch_size=64, epochs=10, verbose=2)

model.evaluate(x_test, y_test, batch_size=64, verbose=1)
````

# Adding Regularization with L2 and dropout

regularization = method to reduce overfitting. To avoid overfitting you could also use L2 Regularization, dropout, data argumentaions etc.

This model now incorporates both L2 regularization in the weights and dropout for regularization.

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import cifar10
from tensorflow.keras import regularizers


physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices: # taki gpu na hone pr error na aaye : array index out of the range
  tf.config.experimental.set_memory_growth(physical_devices[0], True)

(x_train, y_train), (x_test, y_test) = cifar10.load_data()

x_train = x_train.astype("float32")/255.0
x_test = x_test.astype("float32")/255.0


# new : adding L2 Regularization on each conv2d layers

def my_model():
    inputs = keras.Input(shape=(32, 32, 3))
    x = layers.Conv2D(32, 3, padding='same', kernel_regularizer=regularizers.l2(0.01))(inputs)
    x = layers.BatchNormalization()(x)
    x = keras.activations.relu(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Conv2D(64, 3, padding='same', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.BatchNormalization()(x)
    x = keras.activations.relu(x)
    x = layers.Conv2D(128, 3, padding='same', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Flatten()(x) # don't forget the flatten layer
    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)  # fc layer pr bhi regularizer laga do
    # new dropout layer in between weights
    x = layers.Dropout(0.5)(x) # hafl of the connections b/w just above and just below written layers
    outputs = layers.Dense(10, activation='softmax')(x)  # Use softmax activation for multiclass classification, from_logit ko false hi rhne dete hain is bar, softmax yhan use kr lete hain

    model = keras.Model(inputs=inputs, outputs=outputs)
    return model

model = my_model()

model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(),
    optimizer=keras.optimizers.Adam(learning_rate=3e-4),
    metrics=['accuracy']
)

model.fit(x_train, y_train, batch_size=64, epochs=150, verbose=2)

model.evaluate(x_test, y_test, batch_size=64, verbose=1)

# gonna take longer to train, because of dropout. Here batchNormalization is also kind of helping along with l2 and dropout in reducing overfitting. So run for 150 epochs now
````
