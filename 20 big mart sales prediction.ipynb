we're trying to fill missing values in the "Outlet_Size" column of a DataFrame called big_mart_data based on the mode of each "Outlet_Type".

```python
# Calculate mode of Outlet_Size for each Outlet_Type
mode_of_Outlet_size = big_mart_data.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0])) # (lambda "input" : return "some operaations on x" )
```

```python
mode_of_Outlet_size
```

big_mart_data.pivot_table(): This part of the code uses the pivot_table() function provided by pandas. It is a powerful tool to summarize and aggregate data. It rearranges the data in a DataFrame, providing a new shape and structure for the data.

values='Outlet_Size': This parameter specifies the column whose values will be used in the pivot table. Here, it's 'Outlet_Size'. The pivot table will be constructed using the values in this column.

columns='Outlet_Type': This parameter specifies the column whose unique values will become the columns in the pivot table. Each unique value in the 'Outlet_Type' column will become a separate column in the resulting pivot table.

(lambda "input" : return "some operaations on x" )

aggfunc=(lambda x: x.mode()[0]): This parameter specifies the aggregation function to be applied when there are multiple values for a specific combination of 'values' and 'columns'. Here, a lambda function is used. A lambda function is a small, anonymous function defined using the keyword lambda. In this case, it calculates the mode (most common value) for each group and selects the first mode using [0].

Let's say we have a group x with 'Outlet_Size' values: ['Small', 'Medium', 'Small', 'Large'].

x.mode() would return a Series: ['Small'] because 'Small' is the most common value.
x.mode()[0] selects the first (and only) element of the Series, which is 'Small'.
This process ensures that, in case of ties (multiple modes), the code picks one of the modes (the first one encountered in the list of modes). Choosing a single mode is important when filling missing values because it ensures consistency - every missing value for a particular 'Outlet_Type' gets filled with the same 'Outlet_Size'.

Putting it all together, the pivot_table() function creates a table where each column represents a unique 'Outlet_Type', and the values in those columns are the modes of 'Outlet_Size' corresponding to each 'Outlet_Type'. The result is stored in the mode_of_Outlet_size variable.


lambda x: x.mode()[0]


x is a variable that represents each group of data created by the pivot_table() function.
The pivot_table() function automatically groups the data by unique combinations of 'Outlet_Type' values, and for each group, the lambda function is applied to the 'Outlet_Size' column within that specific group.

In simpler terms, x represents each unique 'Outlet_Type' group. The lambda function is applied independently to each of these groups.

```python
# Find missing values in Outlet_Size column
miss_values = big_mart_data['Outlet_Size'].isnull()
```

```python
miss_values
```

```python
# Fill missing values based on Outlet_Type
big_mart_data.loc[miss_values, 'Outlet_Size'] = big_mart_data.loc[miss_values, 'Outlet_Type'].apply(lambda x: mode_of_Outlet_size[x])
```

big_mart_data.loc[miss_values, 'Outlet_Size']: This part of the code selects rows in the 'Outlet_Size' column where the values are missing (NaN) based on the boolean mask miss_values.

= big_mart_data.loc[miss_values, 'Outlet_Type'].apply(lambda x: mode_of_Outlet_size[x]): For these selected rows, this part of the code applies a lambda function to each value in the 'Outlet_Type' column. It maps each 'Outlet_Type' to its corresponding mode of 'Outlet_Size' calculated in the mode_of_Outlet_size pivot table.

lambda x: mode_of_Outlet_size[x]: This lambda function takes an 'Outlet_Type' value as input and returns the mode of 'Outlet_Size' corresponding to that 'Outlet_Type'.

The result of this operation is that missing values in the 'Outlet_Size' column are filled based on the mode of 'Outlet_Size' for their corresponding 'Outlet_Type'.

```python
X = big_mart_data.drop(columns='Item_Outlet_Sales', axis=1)
Y = big_mart_data['Item_Outlet_Sales']
```

```python
# altough test data is separately given, but then usme alag se label encoding and preprocessing etc krna padega, So isi data ka partition kr lete hain
x_train, x_test,y_train, y_test = train_test_split(X,Y,test_size=0.3, random_state=4)
# more data me 70 30 is better.
# In most cases, a 70% training and 30% testing split is better than a 50% training and 50% testing split.
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

XGBoost regressor model training

```python
regressor=XGBRegressor()
```

```python
regressor.fit(x_train, y_train)
```

```python
# so model is trained now.
# we can use this model named "regressor" to predict y given x
```

Evaluation

```python
# R suare values :
r2_train = metrics.r2_score(y_train, training_data_prediction)
r2_train # generally r2 close to 1 is better model, but not always. It is relative. More or less hona we cannot say model performance is better or worse
```

```python
# Predivtion for testing data
```

```python
r2_test = metrics.r2_score(y_test, test_data_prediction)
r2_test
```
