Random Forest Model: Random Forest is a machine learning algorithm that can be used for both regression and classification tasks. It works by constructing multiple decision trees during training and outputs the average prediction (for regression problems) or the mode of the classes (for classification problems) of the individual trees. Your explanation of randomly choosing samples and averaging the outcomes is correct, as Random Forest builds diverse trees by using random subsets of the data for each tree and combines their predictions to improve accuracy and reduce overfitting.

However, there's a bit more to it:

Ensemble Learning: Random Forest is an example of ensemble learning, where multiple models are combined to make a final prediction. The "wisdom of the crowd" principle is applied here: by combining multiple models, the overall performance can be better than that of individual models.

Bagging: Random Forest also utilizes a technique called bagging (Bootstrap Aggregating). It creates multiple subsets of the dataset with replacement (bootstrap samples) and builds a separate tree for each subset. These subsets introduce diversity among the trees, leading to a more robust and accurate model.

Averaging and Voting: For regression problems, the predictions from individual trees are averaged to get the final prediction. In the case of classification problems, a majority vote is taken among the predictions of individual trees.

Data Preprocessing

```python
# separating data and labels
X=df.drop('quality',axis=1)
```

```python
X
```

```python
#Label Binarization
```

```python
# I want quality to be good and bad, instead of having 5 or 6 or more qualities
```

```python
# 0-bad, 1-Good quality
Y=df['quality'].apply(lambda y_value: 1 if y_value >=7 else 0) # lambda function ka use
# k/a label binarization or label encoding
```

```python
Y
```

```python
Y.value_counts()
```

Train test split


```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=2, stratify=Y)
```

```python
print(X.shape, y_train.shape, y_test.shape)
```

Random forest classifier

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

Model evaluation by accuracy score as again classification problem

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
training_data_accuracy
```

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
testing_data_accuracy
```

# Pretty good accuracy by random forest classifier

Building a Predictive System

```python
input_data=(7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)

# changing the input data to numpy array
input_data_As_numpy_array=np.asarray(input_data)

# reshape the 1D arrya as in form of Table as we are predicting just one
input_data_reshaped=input_data_As_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

# just for beautification in output
if(prediction[0]==1):
  print("Good Quality Wine")
else :
  print("Bad Quality Wine")
```
