```python
# importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
# importing the models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# data collection and preprocessing
heart_data = pd.read_csv('/content/heart.csv')
```

```python
# splitting the features and target
```

```python
X = heart_data.drop(columns='target', axis = 1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

## Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y, random_state = 3)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

# comparing the performance of the model

```python
# list of models
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(),RandomForestClassifier() ]
```

```python
def compare_models_train_test():

  for model in models:

    # training the model
    model.fit(x_train, y_train)

    # evaluating the model
    test_data_prediction = model.predict(x_test) # inbuilt knn of sklearn is also similiar to all other sklearn models. Although working is like our built from scratch code.

    accuracy = accuracy_score(y_test, test_data_prediction)

    print('Accuracy score of the ',model, ' = ',accuracy)
```

```python
compare_models_train_test()
```

```python
# theoretically Random forest should be best in accuracy and KNN should be worst in accuracy.
# for different values of random states = different model accuracies.
# so cross validation comes in picture.
# stratifiedKfold is always better than KFold. because it splits the datasets in equal proportion of labels
# Cross validation is used for model comparison.
```

# cross validation. is always better than train test split

it will carry out train test split  everything k times

```python
# cv = cross validation, lr = logistic regression
cv_score_lr = cross_val_score(LogisticRegression(max_iter=1000), X, Y, cv=5) # cv=5 => automatically uses stratified k forld. only kfold use krne se data won't be stratified
# cv = 5 => model will be trained for 5 iterations
# ek bat for logistic regression : converse nhi ho rha 100 iterations me, because standardized nhi hai features. Either do standardization or increase number of iterations
```

```python
print(cv_score_lr) # all 5 splits ke accuracy_score
# take mean accuracy = that will be the original accuracy
mean_accuracy_lr = sum(cv_score_lr)/len(cv_score_lr)
mean_accuracy_lr = mean_accuracy_lr * 100
mean_accuracy_lr = round(mean_accuracy_lr,2) # 2 decimal places tk rounding off kr dega
print(mean_accuracy_lr)
```

svc ke liye k fold cross validation


```python
cv_score_svc = cross_val_score(SVC(kernel='linear'), X, Y, cv=5)

print(cv_score_svc) # all 5 splits ke accuracy_score

# take mean accuracy = that will be the original accuracy
mean_accuracy_svc = sum(cv_score_svc)/len(cv_score_svc)
mean_accuracy_svc = mean_accuracy_svc * 100
mean_accuracy_svc = round(mean_accuracy_svc,2) # 2 decimal places tk rounding off kr dega
print(mean_accuracy_svc)

# clearly svc and logistic regression ki accuracies are same
```

# function to automate the things. All models ka comparison

```python
# list of models
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(),RandomForestClassifier() ]
```

```python
def compare_models_cross_validation():

  for model in models:

    cv_score = cross_val_score(model, X, Y, cv = 5)

    mean_accuracy = round( (sum(cv_score)*100) / (len(cv_score)) ) # lets round to integer values

    print('Cross validation accuracies for the ',model,' = ',cv_score)

    print("Accuray % of mean accuracy for the above model", mean_accuracy)

    print("********************************************************")
```

```python
compare_models_cross_validation()
```
