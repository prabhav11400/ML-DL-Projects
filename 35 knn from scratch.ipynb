 K-Nearest Neighbours ( KNN )
 1. Supervised Learning Model
 2. Used for both classification and Regression
 3. Can be used for non linear models
 4. new data point : new centre : se k neighbours count krega, majority class category will be assigned to the new datapoint

 k should be odd. This is all knn classifier


 **knn regressor â‰**

 k nearest datapoints ki values ka average will be value of the new unknown datapoint

```python
# Advantages of knn
```

1. works well with smaller datasets with less number of features
2. easy to implement for multi-class classification problems
3. Different distance criteria can be used eg: Euclidian distance, Manhattan Distance

```python
# Disadvantages of knn
```

1. Chosing optimum k value
2. Less efficient with high dimensional data
3. Does not perfom well on imbalanced datasets
4. Sensitive to outliers

1. Euclidian Distance  

sqrt((x2-x1)^2 + (y2-y1)^2) : shortest distance between A and B

2. Manhattan Distance

d= | x2-x1 | + |y2-y1 |


Manhattan distance is preferred over Euclidian distance when there is **high dimensionality** in data. as it can deal with complex directions, that eucledian one cannot do.

```python
# neeche knn code ko bde patience ke sath padhe, samjh aa jayega. Patience rakhiyega. This code is completely different from the implementations of other ML models
# abhi se yad rkho :  distance_list = [] is list of tuples. Na ki list of 1,2,3 like this. each tuple is like = ( (f1,f2,f3,f4,....,fn,target), d ) where d is distance of test_data to that data_point jiske all features are written in (f1,f2,f2..,fn)
# and Most Important : hm distance ke rup me : test_data ke features f_i ki available datapoint ke respective feature f_i se distance nikalte hain(for all features), ye koi physical distance nhi hoti hai.
# aur ek bat, X_TRAIN input me bhi features and labels dono de dena hai, features ka use upar ke 3 functions me hoga( jo len -1 lagakar manage ho jayega) nad label column ka use 4th function me hoga( jo -1 indexing ke dvara use me liya jayega)
import numpy as np
import statistics

class KNN_Classifier():

  # Initiating the parameters
  def __init__(self, distance_metric):
    self.distance_metric = distance_metric



  # Getting the distance metric
  # get_distance_metric just gives the distance between two points : available options are : euclidean and manhattan only : [ remember here x_i's are all the features. This distance is mutidimensional and this this not the actual physical distance, but distane between points having each coordinate as one feature value]

  def get_distance_metric(self, training_data_point, test_data_point):
    if (self.distance_metric == 'euclidean'):
      dist = 0
      # Calculate Euclidean distance between two data points
      for i in range(len(training_data_point) - 1): # -1, since target column ko x_i ke rup me include nhi krna hai, only features has to be dimensions for being included in calulating the distance
        dist = dist + (training_data_point[i] - test_data_point[i])**2
      euclidean_dist = np.sqrt(dist)
      return euclidean_dist
    elif (self.distance_metric == 'manhattan'):
      dist = 0
      # Calculate Manhattan distance between two data points
      for i in range(len(training_data_point) - 1): # again -1, since last column label hai , use include nhi krna as a feture for distance calculation
        dist = dist + abs(training_data_point[i] - test_data_point[i])
      manhattan_dist = dist
      return manhattan_dist



  # Getting the nearest neighbors
  def nearest_neighbors(self, X_train, test_data, k): # remember 'test_data' is single point, not equivalent to y_train. Isi point ko hme classify krna hai.
    distance_list = [] #an array of tuples, whose first element(0th index) is the data point and second element(1st index) is the distance of that training datapoint to the unknown new datapoint(i.e. test_data)

    # Calculate distances from the test_data point to all training data points : test data is only one, jiski jankari hme leni hai ki vo kis class me lie karega.keval usi ek point ki distance all other points se nikalana hai
    for training_data in X_train: # for all elements in x_train ( x_train me sare features hain, with the label, label ka neeche prediction me use hai, distance calculation me -1 lagake unhe exclude kr diya gya hai)
      distance = self.get_distance_metric(training_data, test_data)
      distance_list.append((training_data, distance)) # hr ek point se, us unknown point ki jo distance aayegi, that will be appended in distance_list array in form of tuple.
      # sahi se dekho, distance_list me tuple is appended, not single distance as an element. Keval single distances store krenge to ptha kaise chalega after sorting ki kis datapoint ki distance thi vo.
      # at the end hme vo datapoints hi return krne hain, bad me un deta points ke labels ki analysis hogi, and max number vali category hamare test_data ko assign ho jayegi

    # Sort the distances and select the top k neighbors
    distance_list.sort(key=lambda x: x[1])
    # This line of code sorts the distance_list based on the second element (x[1]) of each element (a tuple) in the list

    # ab in top k distance ke corresponding datapoints ko hi as a output dene ke liye, ek aur array banao and first k tuples ke first element i.e. data_point ko store krke return kr do
    # obviously data_point means all the values of its features
    neighbors_list = [] # that will be list of datapoints = assume it as 2D array, because it is going to be list of 1D arrays => effectively a table bol skte hain
    for j in range(k): # see for loop keval k times hi chalega = k nearest neighbors me padne vale hi datapoints are being stored.
      neighbors_list.append(distance_list[j][0]) # first element of each tuple, i.e. datapoint from the sorted distance_list array is now being stored in neighbour_list array
    return neighbors_list # so finally k nearest neighbours ke sbhi datapoints are returned



  # Predict the class of the new data point
  def predict(self, X_train, test_data, k):
    neighbors = self.nearest_neighbors(X_train, test_data, k) # neighbors array will get to store the k nearest datapoints of the new(test_data) datapoint

    # Create an empty list to store the labels of these k points
    label = []

    for data in neighbors:
      # Append the last element (class label) of each neighbor to the label list : assuming ki datapoint ka last column hi labels/target ko store krta hai
      label.append(data[-1]) # -1 => negative indexing = last column # here we are using the labels, only one place pr. for all rows in neighbours 2D array, store their last elements in the label array

    # Find the most common class label among the neighbors
    predicted_class = statistics.mode(label)

    return predicted_class


```

1. distance_list.sort(key=lambda x: x[1]):

 This line of code sorts the distance_list based on the second element (x[1]) of each element (a tuple) in the list. In this context, distance_list contains tuples where the first element of each tuple is a training data point, and the second element is the corresponding distance from the test data point. Sorting by x[1] means that the list will be sorted based on the distances.

2. distance_list[j][0]: After sorting distance_list, this code retrieves the training data point from the j-th element of the sorted list. The [j] index selects the j-th element of the list, and the [0] index within that element retrieves the first element of the tuple, which is the training data point. This code is used to extract the k nearest neighbors from the sorted list.

so remember distance_list is an array of tuples, whose first element(0th index) is the data point and second element(1st index) is the distance of that training datapoint to the unknown new datapoint(i.e. test_data)
