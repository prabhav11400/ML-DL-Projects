Very unbalanced dataset

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head(10)
```

```python
credit_card_data.tail(10)
```

```python
credit_card_data.info()
```

```python
# Checking missing values in each columns
credit_card_data.isnull().sum()
# No missing value
```

```python
# Distribution of legit andfraudulent transections
credit_card_data['Class'].value_counts()
# only 0.3% belong to 2nd class. So highly imbalanced dataset
```

```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape, fraud.shape)
```

```python
# Statistical Measures
legit.Amount.describe()
```

```python
fraud.Amount.describe() # used to describe only one column's statistics
```

```python
# copare the values for the transections
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

```python
# legit se 492 datapoints ko randomly lekar concatenate krne ja rha
```

```python
legit_sample=legit.sample(n=492)
```

```python
df=pd.concat([legit_sample,fraud],axis=0) # axis = 0 means row wise concatenate in panda, Numpy me ulta
```

```python
df.head()
```

```python
df['Class'].value_counts()
```

```python
df.groupby('Class').mean()
# almost similar as with all datapoints
```

Features and Labels(targets) extraction

```python
X = df.drop(columns='Class',axis=1)
Y = df['Class']
```

Train test splitting

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

Model Training Logistic Regression

```python
model=LogisticRegression()
```

```python
# training the logistic regression model with our training data
model.fit(x_train,y_train)
```

Model Evaluation

```python
# Accuracy score
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
training_data_accuracy
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
test_data_accuracy
```
